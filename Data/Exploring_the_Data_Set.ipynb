```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = food['cuisine'].value_counts().index
sizes = [food['cuisine'].value_counts()[i] for i in range(20)]
explode = [i/30 for i in range(20)]

#fig1, ax1 = plt.subplots()
#ax1.pie(sizes, labels=labels, #autopct='%1.1f%%',
#        shadow=True, startangle=90)
#ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

fig = plt.figure(figsize =(8, 14))
plt.pie(sizes, 
        labels = labels, 
        autopct='%1.1f%%', 
        textprops={'fontsize': 22}, 
        explode=explode, 
        startangle=35, 
        shadow=True)

#plt.rcParams.update({'font.size': 22})
plt.title('Cuisines', fontsize=30)

plt.show()
```

```python
## first entry is the desired list that will be searched
    ## duplication of words in list is allowed
## second entry is the letters/characters that will be searched for 
    ## Must be entered in form 'letter'
        ## for example, to search for the word onion, use 'onion'
## third entry is either the word True or the word False
    ## True produces a list
    ## False prints the length and prints the list
    
## Remark: to use the search function, get a list of all of the ingredients first - do not run from a dataframe

def search(the_list,word,is_list):
    mylist = []
    myword = word
    
    for element in the_list:
        if myword in element:
            mylist.append(element)
    
    series_version = pd.Series(mylist)
    
    if is_list == True:
        return(mylist)
    
    if is_list == False:
        print("Lenth of list is", len(series_version.value_counts().index))
        print()
        print("Elements from the list containing the characters:")
        print(series_version.value_counts().index)
```

```python
search(tot_ingred_series.value_counts().index, 'food colo', False)
```

```python
lengths = food['ingredients'].apply(len)
```

```python
fig = plt.figure(figsize =(8, 6))

sns.histplot(lengths, bins = 65)

plt.xlabel("Number of Ingredients", fontsize=22)
plt.ylabel("Number of Recipes", fontsize=22)

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.title("Ingredient Count Distribution", fontsize=25)

plt.show()
```

```python
lengths.describe()
```

```python
for i in range(len(lengths)):
    if lengths[i] == lengths.max():
        print("The longest recipe comes from:", food['cuisine'][i])
        print()
        print("Its ingredient list has", lengths[i], "ingredients")
        print()
        print("Ingredient list:", food['ingredients'].iloc[i])
```

```python
## Shortest recipes

j = 1
for i in range(len(lengths)):
    if lengths[i] == lengths.min():
        print("Recipe number", j)
        print()
        print("From:", food['cuisine'][i])
        print()
        print("Ingredient list:", food['ingredients'].iloc[i])
        print()
        print()
        j = j+1
```
